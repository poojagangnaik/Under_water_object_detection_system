```python
import cv2
import pandas as pd
from ultralytics import YOLO
import cvzone
import numpy as np
from time import sleep
import serial

print("Libraries Updated")
```

```python
# Load COCO class names
with open("coco9label.txt", "r") as f:
    class_names = f.read().splitlines()

# Load the YOLOv8 model
# model = YOLO("yolo11n.pt")   # nano model preferred for Raspberry PI
model = YOLO("best9label.pt")   # nano model preferred for Raspberry PI

# Open the video file (use video file or webcam, here using webcam)
cap = cv2.VideoCapture(0)

ser = 0
def init_serial():
    COMNUM = 4          #Enter Your COM Port Number Here.
    global ser          #Must be declared in Each Function
    ser = serial.Serial('com3', 9600)
    #ser = serial.Serial('COM6') # open COM24
    ser.baudrate = 9600 # set Baud rate to 9600
    ser.bytesize = 8    # Number of data bits = 8
    ser.parity   = 'N'  # No parity
    ser.stopbits = 1    # Number of Stop bits = 1
    #time.sleep(5)
    ser.timeout = 10
    if ser.isOpen():
        print( 'Open: ' + ser.portstr)

# comment if hardware is not connected
#init_serial()
###test = "SERIAL TEST"
#ser.write(b'SERIAL TEST')
print("Camera and Serial Done")
```

```python
class_id_old = 5
class_id_count = 0
count=0
while True:
    ret, frame = cap.read()
    
    count += 1
    if count % 3 != 0:
        continue
#    frame=cv2.flip(frame,-1)
    frame=cv2.flip(frame,1)

#    frame = cv2.resize(frame, (1020, 500))
    
    # Run YOLOv8 tracking on the frame, persisting tracks between frames
    results = model.track(frame, persist=True,imgsz=256)

    # Check if there are any boxes in the results
    if results[0].boxes is not None and results[0].boxes.id is not None:
        # Get the boxes (x, y, w, h), class IDs, track IDs, and confidences
        boxes = results[0].boxes.xyxy.int().cpu().tolist()  # Bounding boxes
        class_ids = results[0].boxes.cls.int().cpu().tolist()  # Class IDs
        track_ids = results[0].boxes.id.int().cpu().tolist()  # Track IDs
        confidences = results[0].boxes.conf.cpu().tolist()  # Confidence score
       
        for box, class_id, track_id, conf in zip(boxes, class_ids, track_ids, confidences):
            c = class_names[class_id]
            x1, y1, x2, y2 = box
            cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),2)
            cvzone.putTextRect(frame,f'{track_id}',(x1,y2),1,1)
            cvzone.putTextRect(frame,f'{c}',(x1,y1),1,1)

            print(class_id)
            if( class_id_old == class_id ):
                class_id_count = class_id_count + 1
                print(class_id_count)
                if( class_id_count >= 10 ):
                    print("Material Detected")
                    if( class_id == 0 ):
#                        ser.write(b'#AA@')                        
                        print("#AA@ - battery")
                    elif( class_id == 1 ):
#                        ser.write(b'#BB@')                        
                        print("#BB@ - bottle")
                    elif( class_id == 2 ):
#                        ser.write(b'#CC@')                        
                        print("#CC@ - bread")
                    elif( class_id == 3 ):
#                        ser.write(b'#DD@')                        
                        print("#DD@ - chapati")
                    elif( class_id == 4 ):
#                        ser.write(b'#EE@')                        
                        print("#EE@ - medicine")
                    elif( class_id == 5 ):
#                        ser.write(b'#FF@')                        
                        print("#FF@ - suparipack")
                    elif( class_id == 6 ):
#                        ser.write(b'#GG@')                        
                        print("#GG@ - syringe")
                    elif( class_id == 7 ):
#                        ser.write(b'#HH@')                        
                        print("#HH@ - tetrapack")
                    elif( class_id == 8 ):
#                        ser.write(b'#II@')                        
                        print("#II@ - tomato")
                    class_id_count = 0
                    sleep(2)
                    while(1):
                        if cv2.waitKey(1) & 0xFF == ord("c"):
                           break
            else:
                class_id_count = 0
                    
            class_id_old = class_id
            

    cv2.imshow("RGB", frame)
    sleep(0.3)
                
    if cv2.waitKey(1) & 0xFF == ord("q"):
       break
```

```python
# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()

print("Project End")
```
